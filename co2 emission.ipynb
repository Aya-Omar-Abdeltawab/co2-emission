```python
top_10=dict(df["Country"].value_counts())
for country in top_10.keys():
        top_10[country]=0.0
top_10
```

```python
from collections import Counter
maxnum=int(df.shape[0])
for num in range(0,maxnum):
    top_10[df['Country'][num]]+=df['CO2 Emissions'][num]
top_10

top_values = Counter(top_10).most_common(10)
top_values
```

```python
plt.figure(figsize=(10, 6))

# Extracting countries and CO2 emissions from top_values
countries = [country for country, _ in top_values]
emissions = [emission for _, emission in top_values]

# Creating the horizontal bar plot
plt.barh(countries, emissions, color='skyblue')

plt.xlabel('CO2 Emissions')
plt.ylabel('Country')
plt.title('Top 10 Countries by CO2 Emissions')

plt.show()
```

```python
time=dict(df["TIME"].value_counts())
for TIME in time.keys():
    time[TIME]=0.0

for num in range(0,maxnum):
    time[df['TIME'][num]]+=df['CO2 Emissions'][num]
time
```

```python
top_values = Counter(time).most_common(10)
top_values
```

```python
years = [int(year) if isinstance(year, int) else int(year.split('-')[0]) for year, _ in top_values]
emissions = [emission for _, emission in top_values]

plt.figure(figsize=(10, 6))

# Creating the bar plot
plt.bar(years, emissions, color='skyblue')

plt.xlabel('Year')
plt.ylabel('CO2 Emissions')
plt.title('CO2 Emissions Over Time')
plt.xticks(rotation=45)  

plt.tight_layout()
plt.show()
```

# New feature (climate change impact)

```python
# Define function to calculate climate change impact based on criteria
def calculate_impact(row):
    if  row['Flight type'] == 'All flights' and row ['Frequency']=='Monthly' or row['CO2 Emissions']>100 or row['Country']  in top_10 :
        return 'High Impact'
    elif row['Flight type'] == 'Freight flights' or row ['Frequency']=='Quarterly' :
        return 'Medium Impact'
    else:
        return 'Low Impact'

# Apply the function to create the new column 'climate_change_impact'
df['climate_change_impact'] = df.apply(calculate_impact, axis=1)
```

# Decision Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier as dt
from sklearn import metrics
from sklearn.model_selection import train_test_split, GridSearchCV

# Split dataset into features and target variable
x = df[['Source of emissions', 'Flight type', 'Country', 'TIME', 'Frequency', 'CO2 Emissions']]
y = df['climate_change_impact']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.30, random_state=42)

# Initialize and train the Decision Tree model
dt_model = dt(random_state=42)
dt_model.fit(x_train, y_train)

# Print scores
print("Decision Tree Score:", dt_model.score(x_test, y_test))
print("Decision Tree F1 Score:", metrics.f1_score(y_test, dt_model.predict(x_test), average=None))
print("Decision Tree R2 Score:", metrics.r2_score(y_test, dt_model.predict(x_test)))

# Define parameter space for GridSearchCV
parameter_space = {
    'max_depth': [None, 10, 20],  # values for max_depth
    'min_samples_split': [2, 5, 10],  # values for min_samples_split
}

# Initialize GridSearchCV
grid_search = GridSearchCV(dt_model, parameter_space, cv=5)
grid_search.fit(x_train, y_train)

# Get the best hyperparameters
best_params = grid_search.best_params_
print("Best hyperparameters of Decision tree:", best_params)
```

```python
from sklearn.ensemble import RandomForestClassifier
# Initialize and train the Random Forest model
rf = RandomForestClassifier(random_state=42)
rf.fit(x_train, y_train)

# Print scores
print("Random Forest Score:", rf.score(x_test, y_test))
print("Random Forest F1 Score:", metrics.f1_score(y_test, rf.predict(x_test), average=None))
print("Random Forest R2 Score:", metrics.r2_score(y_test, rf.predict(x_test)))
```
